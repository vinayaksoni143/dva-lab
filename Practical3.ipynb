## Least Squares Regression

#### Functions and formulae

y = mx + b

m = N\*sum(x\*y) - sum(x)\*sum(y) / N\*sum(x^2) - sum(x)^2

b = sum(y) - m\*sum(x)/N

```python
# func
def getLine(x, y):
  x_square = [i*i for i in x]
  xy = []
  for i in range(len(x)):
    xy.append(x[i]*y[i])

  sm_x = 0
  for i in x:
    sm_x += i

  sm_y = 0
  for i in y:
    sm_y += i

  sm_x2 = 0
  for i in x_square:
    sm_x2 += i

  sm_xy = 0
  for i in xy:
    sm_xy += i

  N = len(x)

  m = (N*sm_xy - sm_x*sm_y) / (N*sm_x2 - sm_x*sm_x)
  b = (sm_y - m*sm_x) / N

  m = m.round(2)
  b = b.round(2)
  return [m, b]
```

```python
# c: Houses with how many bedrooms are most sold?
d = {}
for i in df["bedrooms"]:
  if i not in d.keys():
    d[i] = 0
  d[i] += 1

res = -1
for i in d.keys():
  if res == -1 or d[i] > d[res]:
    res = i
print("Houses with how many bedrooms are most sold?")
print("ans:",res)
```

```python
# (e) Find the correlated features/ columns
from scipy.stats import pearsonr
ls = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']
for i in ls:
  for j in ls:
    if i == j:
      continue
    a = np.array(df[i])
    b = np.array(df[j])
    corr, _ = pearsonr(a, b)
    print(i,"and",j,":",corr.round(4))
```

```python
# (f) Find null values and fill with mean value for all columns
print("no null values")
```

```python
X = df.drop('date',axis= 1)
y = df['price']
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
```

```python
predictions = regr.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error

print('mean_squared_error : ', mean_squared_error(y_test, predictions))
print('mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```
